```python
import pandas as pd
```

```python
df_cars = pd.read_csv("./data/vehiculos_electricos_hibridos.csv")
```

```python
# Reemplazar nombres de las columnas
df_cars.rename(columns={"AÑO_REGISTRO":"ANIO_REGISTRO"},inplace=True)
```

```python
# Muestra información del dataframe
df_cars.info()
```

```python
# df_cars["EJES"] = df_cars["EJES"].astype("int64")
list_to_change = [
    "CILINDRAJE",
    "CAPACIDAD_CARGA",
    "CAPACIDAD_PASAJEROS",
    "PESO",
    "POTENCIA",
    "EJES"
]

for column_name in list_to_change:
    df_cars[column_name] = pd.to_numeric(df_cars[column_name], errors='coerce').fillna(0)
    df_cars[column_name] = df_cars[column_name].astype("int64")
df_cars.info()
```
