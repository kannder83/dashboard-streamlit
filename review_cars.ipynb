```python
import pandas as pd
import altair as alt
```

## Revision de data

Se revisan los datos del archivo csv y se acondicionan para trabajar posteriormente en Streamlit.

```python
df_cars = pd.read_csv("./data/vehiculos_electricos_hibridos.csv")
```

```python
# Reemplazar nombres de las columnas
df_cars.rename(columns={"AÑO_REGISTRO":"ANIO_REGISTRO"},inplace=True)
```

```python
# Retirar los NaN de una columna:
df_cars["MODALIDAD"].fillna("", inplace=True)
```

```python
# Se quitan los nan y se reemplaza de float por integer.
list_to_change = [
    "CILINDRAJE",
    "CAPACIDAD_CARGA",
    "CAPACIDAD_PASAJEROS",
    "PESO",
    "POTENCIA",
    "EJES"
]

for column_name in list_to_change:
    df_cars[column_name] = pd.to_numeric(df_cars[column_name], errors='coerce').fillna(0)
    df_cars[column_name] = df_cars[column_name].astype("int64")

df_cars.info()
```

## Graficos

```python
# Top de marcas de vehiculos más registrados por año

data_line = pd.DataFrame(df_cars.groupby(
        ['ANIO_REGISTRO', "MARCA"]).size().reset_index(name='COUNT'))

sort_df = pd.DataFrame(data_line.sort_values(
        by="COUNT", ascending=False)).reset_index(drop=True)

list_marca = pd.DataFrame(sort_df["MARCA"].unique(), columns=["MARCA"])

list_marca = list_marca.head(10)

sort_df = sort_df[sort_df["MARCA"].isin(list_marca["MARCA"])]

line_chart = alt.Chart(sort_df).mark_bar(opacity=0.8).encode(
        y=alt.Y("MARCA:N", title=""),
        x=alt.X("sum(COUNT):Q", title="Vehiculos Registrados"),
        color=alt.Color("ANIO_REGISTRO:N", title="Año de Registro",
                        scale=alt.Scale(scheme="tableau20")),
        order=alt.Order(
            'COUNT:N',
            sort="descending"
        )
    )

line_chart
```

```python
# Grafica de torta que muestra el uso que se le da a los carros electricos o hibridos registrados.

data_line = pd.DataFrame(df_cars.groupby(
    ["ANIO_REGISTRO", "MUNICIPIO", "COMBUSTIBLE", "SERVICIO"]).size().reset_index(name='COUNT'))

line_chart_left = alt.Chart(data_line).mark_arc(innerRadius=50).encode(
    theta="sum(COUNT)",
    color=alt.Color("COMBUSTIBLE", title="Tipo de Motor",
                    scale=alt.Scale(scheme="tableau10"))
)

line_chart_left
```

```python
line_chart_right = alt.Chart(data_line).mark_arc(innerRadius=50).encode(
    theta="sum(COUNT)",
    color=alt.Color("SERVICIO", title="Tipo de Servicio",
                    scale=alt.Scale(scheme="category10"))
)

line_chart_right
```

```python
# Grafico del tipo de vehiculo registrado por año

data_line = pd.DataFrame(df_cars.groupby(
    ['ANIO_REGISTRO', 'CLASE']).size().reset_index(name='COUNT'))

data_line = data_line.sort_values(by='COUNT', ascending=False)

line_chart = alt.Chart(data_line).mark_bar().encode(
    x=alt.X("sum(COUNT):Q", title="Cantidad de Vehiculos"),
    y=alt.Y("ANIO_REGISTRO:N", title=""),
    color="CLASE:N",
    order=alt.Order(
          'COUNT:N',
          sort="descending"
    )
)

line_chart
```
